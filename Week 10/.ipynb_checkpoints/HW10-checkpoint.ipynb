### 1.
**(a)**
It is linear:
$$\begin{align*}(f+h, g)_E &= \int_0^1p(x)[f'(x)+h'(x)]g'(x)dx+\int_0^1r(x)[f(x)+h(x)]g(x)dx\\
&=\int_0^1p(x)f'(x)g'(x)dx+\int_0^1r(x)f(x)g(x)dx + \int_0^1p(x)h'(x)g'(x)dx+\int_0^1r(x)h(x)g(x)dx\\
&= (f,g)_E + (h,g)_E
\end{align*}$$

Also, it is obvious that $(f, g)_E = (g, f)_E$ and $(kf, g)_E = k(f, g)_E$. The last requirement is
$$(f, f)_E\ge 0\Longleftrightarrow \int_0^1p(x)f'(x)^2dx+\int_0^1r(x)f(x)^2dx\ge 0$$
which is true as $p(x), r(x) > 0$ and $f'(x)^2, f(x)^2\ge 0$.

**(b)** Note that
$$\int_0^1mf'(x)^2dx \le \int_0^1p(x)f'(x)f'(x)dx \le \int_0^1Mf'(x)^2dx$$
and similarly
$$\int_0^1mf(x)^2dx \le \int_0^1r(x)f(x)f(x)dx \le \int_0^1Mf(x)^2dx$$
so
$$m||f||^2_{1,2} = \int_0^1mf'(x)^2dx+\int_0^1mf(x)^2dx\le ||f||^2_E \le \int_0^1Mf'(x)^2dx +\int_0^1Mf(x)^2dx = M||f||^2_{1, 2}$$

**(b)**
It approaches a straight line:

($a=100$)

Why? The homogenous equation is
$$x^2-ax-2 = 0\implies x=\frac{a\pm\sqrt{a^2+8}}{2}$$
when $a$ becomes very large the smaller root approaches $0$ and the larger root approaches $a$. Notice that for very small values of $k$
$$k(e^{kx}-1)\approx x$$
so the smaller root will assume a close to linear function. The solution will approximate the function
$$c_1x + c_2e^{ax}$$
But the boundary condition $u(1)=0$ gives
$$c_2=-\frac{c_1}{e^a}\approx 0$$
and so it vanishes, and we are left with just $c_1x.$

### 3.
**(a)**
We have
$$u''(x_k)\approx \frac{u(x_{k+1}) - 2u(x_{k}) + u(x_{k-1})}{h^2}$$
and also
$$u'(x_k)\approx \frac{u(x_{k+1})-u(x_{k-1})}{2h}$$

So the difference equations are
$$\left(\frac2{h^2}+r(x)\right)u_k - \frac{u_{k+1} + u_{k-1}}{h^2} = f(x_k)$$
with boundary conditions
$$u_0 = u_n,$$
$$u_1 - u_{-1} = u_{n+1} - u_{n-1}.$$

The matrix $A$ is nearly tridiagonal, except for the last two rows from the boundary conditions.

```python
# (b)
# Coded in Python3 with numpy

import numpy as np

def r(x):
    return (2*np.pi*np.cos(2*np.pi*x))**2

def f(x):
    return 4*np.pi**2*np.exp(np.sin(2*np.pi*x))*np.sin(2*np.pi*x)

def solve(n):
    h = 1/n

    A = []
    b = []

    # First boundary cond.
    row = np.zeros(n+3)
    row[1] = 1; row[-2] = -1
    A.append(row); b.append(0)

    # Second boundary cond.
    row = np.zeros(n+3)
    row[0] = -1; row[2] = 1; row[-1] = -1; row[-3] = 1
    A.append(row); b.append(0)

    # Middle points:
    for i in range(0, n+1):
        row = np.zeros(n+3)
        row[i] = -n**2
        row[i+1] = 2*n**2 + r(i*h)
        row[i+2] = -n**2
        A.append(row); b.append(f(i*h))

    A = np.array(A)
    u = np.linalg.inv(A) @ b
    return u

def exact(n):
    return np.exp(np.sin(2*np.pi*np.linspace(0, 1, n+3)))

def L2(approx, target):
    return np.max((approx - target)**2)

n = 12
print("h = 1/12,")
print("L2 error =", L2(solve(n), exact(n)))
```

```python
import matplotlib.pyplot as plt

errors = []
for i in range(3, 11):
    n = 2**i
    errors.append(L2(solve(n), exact(n)))
plt.plot(range(3, 11), errors)
plt.yscale('log')
plt.show()
```

**(c)**
By Taylor's theorem,
$$u(x+h) = u(x) + hu'(x) + \frac{h^2}{2}u''(x) + \frac{h^3}{6}u'''(x) + \frac{h^4}{24}u^{(4)}(x) + O(h^5)$$
so
$$u(x+h) -2u(x) + u(x-h) = h^2u''(x) + \frac{h^4}{12}u^{(4)}(x) + O(h^6).$$

Therefore,
$$\frac{u(x+h) -2u(x) + u(x-h)}{h^2} = u''(x) + \frac{h^2}{12}u^{(4)}(x) + O(h^4).$$

-----
Similarly$^†$,
$$F(x_{k+1}, u(x_{k+1})) = F(x,u)+h\cdot\frac{dF}{dx}+\frac{h^2}{2}\cdot\frac{d^2F}{dx^2}+O(h^3)$$
so
$$\frac1{12}\left[F(x_{k-1}, u(x_{k-1}))+10F(x_{k}, u(x_{k}))+F(x_{k+1}, u(x_{k+1}))\right] = F(x,u)+\frac{h^2}{12}F''(x, u) + O(h^4)$$

Then,
$$-\frac{u(x+h)-2u(x) + u(x-h)}{h^2} + \frac1{12}\left[F(x_{k-1}, u(x_{k-1}))+10F(x_{k}, u(x_{k}))+F(x_{k+1}, u(x_{k+1}))\right] = -u''(x) + F(x, u) + O(h^4) = O(h^4)$$

Expanding out $F$ results in the given equation.

[†Inspiration](http://www.physics.drexel.edu/~steve/Courses/Comp_Phys/Physics-305/numerov.pdf)


```python
# (d)
# Coded in Python3 with numpy

def r(x):
    return (2*np.pi*np.cos(2*np.pi*x))**2

def f(x):
    return 4*np.pi**2*np.exp(np.sin(2*np.pi*x))*np.sin(2*np.pi*x)

def numerov(n):
    h = 1/n

    A = []
    b = []

    # First boundary cond.
    row = np.zeros(n+1)
    row[0] = 1; row[-1] = -1
    A.append(row); b.append(0)

    # Second boundary cond.
    row = np.zeros(n+1)
    row[0] = -1; row[1] = 1; row[-1] = -1; row[-2] = 1
    A.append(row); b.append(0)

    # Middle points:
    for i in range(1, n):
        row = np.zeros(n+1)
        row[i-1] = -n**2 + 1/12*r(i*h-h)
        row[i] = 2*n**2 + 5/6*r(i*h)
        row[i+1] = -n**2 + 1/12*r(i*h+h)
        A.append(row); b.append(1/12*(f(i*h-h) + 10*f(i*h) + f(i*h+h)))

    A = np.array(A)
    u = np.linalg.inv(A) @ b
    return u

def exact(n):
    return np.exp(np.sin(2*np.pi*np.linspace(0, 1, n+1)))

def L2(approx, target):
    return np.max((approx - target)**2)

n = 12
print("h = 1/12,")
print("L2 error =", L2(numerov(n), exact(n)))
```

```python
errors = []
for i in range(3, 11):
    n = 2**i
    errors.append(L2(numerov(n), exact(n)))
plt.plot(range(3, 11), errors)
plt.yscale('log')
plt.show()
```
